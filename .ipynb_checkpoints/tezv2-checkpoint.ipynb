```python
def statusUser(myList,user):
    status=False
    for i in myList:
        if i[0] == user:
            status =True
            break
    return status
def findVerticeAndEdge(myData):
    myVerticesList = []
    myEdgeList=[]
    #count=0
    for i in myData:
        #count = count+1
        #print(count)
        retweetInfos = i.split(",")
        if not statusUser(myVerticesList,retweetInfos[0]):
            infoTupleUser = (retweetInfos[0],retweetInfos[0])
            myVerticesList.append(infoTupleUser)
        if not statusUser(myVerticesList,retweetInfos[1]):
            myVerticesList.append((retweetInfos[1],retweetInfos[1]))
        myEdgeList.append((retweetInfos[0],retweetInfos[1]))
        #print(retweetInfos[0])
    return myVerticesList,myEdgeList
```

```python
myNode = []
myEdge = []
```

```python
step = 1000
items_count = 1000
for j in range(0,3001,step):
    print("j = "+ str(j))
    print("xxxx")
    a= j+step
    for i in range(j,a):
        first = i*step
        last = (i+1)*step
        print(first,last)
        if(last>2500000):
        #if(last>2500000):   
            break
        vertices,edges = findVerticeAndEdge(file[first:last])
        #for x in vertices:
            #print(x)
        myNode = list(set(myNode+vertices))
        myEdge = list(set(myEdge+edges))
```
