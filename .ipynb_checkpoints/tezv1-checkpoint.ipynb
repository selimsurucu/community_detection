```python
from ast import literal_eval
from pyspark import SparkConf, SparkContext
import sys, json
from pyspark import SQLContext
from graphframes import *

conf = SparkConf().setAppName("Graphx")
sqlContext = SQLContext(sc)
```

```python
file = sc.textFile('data\\myData')
```

```python
def mapTweet(line):
  retweetList = []

  myJson = json.dumps(line, ensure_ascii=False)
  l =  literal_eval(myJson)
  try : 
    tweet = json.loads(l)
    retweetList.append([tweet.get('user').get('id_str') + "-" + tweet.get('user').get("screen_name")+ ","+ tweet.get('retweeted_status').get('user').get('id_str')+ "-" + tweet.get('retweeted_status').get('user').get("screen_name"), 1])
    return retweetList
  except:
    return retweetList
```

```python
mapTweets = file.flatMap(lambda line: mapTweet(line))
reducePairUserRetweets = mapTweets.reduceByKey(lambda a,b:a+b).collect()

for i in reducePairUserRetweets[:5] : 
  print(i)
```

```python
def statusUser(myList,user):
    status=False
    for i in myList:
        if i[1] == user:
            status =True
    return status
def findVerticeAndEdge(myData):
    myVerticesList = []
    myEdgeList=[]
    for i in myData:
        myList=[]
        myUsersInfo = i[0].split(",")
        for userInfo in myUsersInfo:
            infos = userInfo.split("-")
            if not statusUser(myVerticesList,infos[0]):
                infoTupleUser = (infos[0],infos[1])
                myVerticesList.append(infoTupleUser)
            myList.append(infos[0])
        myEdgeList.append((myList[0],myList[1]))
    return myVerticesList,myEdgeList
```

```python
vertices,edges = findVerticeAndEdge(reducePairUserRetweets)
```

```python
x = 0
for i in vertices:
    x = x+1
    print(x)
    print(i)
```

```python
for i in edges:
    print(i)
```

```python
v = sqlContext.createDataFrame(vertices, ["id", "name"])
```

```python
e = sqlContext.createDataFrame(edges, ["src", "dst"])
```

```python
g = GraphFrame(v, e)
```

## Triangle Count

```python
results = g.triangleCount()
tc_df=results.select('count').groupBy('count').count().toPandas()
tc_df.columns=['triangle_count','count']
```

```python
tc_df
```

## Label Propagation

```python
result = g.labelPropagation(maxIter=5)
```

```python
tc_df=result.select('label').groupBy('label').count().orderBy("count").toPandas()
tc_df.columns=['label','count']
tc_df
```

## Strongly Connected Components

```python
result = g.stronglyConnectedComponents(maxIter=10)
tc_df=result.select('component').groupBy('component').count().orderBy("count").toPandas()
tc_df.columns=['component','count']
tc_df
```
