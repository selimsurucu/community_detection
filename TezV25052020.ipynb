```python
import networkx as nx
import cdlib
from cdlib import algorithms
from cdlib import evaluation
import numpy as np
from matplotlib import pyplot as plt
```

```python
print('reading retweet edges from file...')

G = nx.DiGraph() 
filename = 'data\\retweet.txt'
fhand = open(filename)
count = 0
for line in fhand:
    lst = line.strip().split(',')
    if len(lst[0])==0 or len(lst[1])==0:
        continue
    
    if count > 350000:
        break
    count = count+1
    
    G.add_weighted_edges_from([(lst[0],lst[1],int(lst[2]))])
```

## Utils

```python
def community_size(com):
    #print("Community_size = " +str(len(com)))
    #print("Max element_count = " + str(len(com[0])))
    return len(com[0]), len(com)
```

```python
def found_com_relation(com):
        com_count = 30
        new_matrix = np.zeros((com_count,com_count))
        for i in range(0,com_count):
            
            users = com[i]
            for user in users:
                for j in range(0,com_count):
                    if i == j :
                        continue
                    for o_user in com[j]:
                        if(G.get_edge_data(user,o_user) != None) : 
                            #print(i,j,new_matrix[i][j])
                            new_matrix[i][j]  = int(new_matrix[i][j]) +1 
                            print(i,j,new_matrix[i][j])
            print("xxxxxxxxxxxxx")
        return new_matrix
            
```

```python

def writeToFile(wr,size):
    f = open(r"myCom"+str(size)+".txt","w") 
    for i in wr:
        f.write(i+"\n")   
    f.close()
```

## LPA

```python
lp_coms = algorithms.label_propagation(G.to_undirected())
```

```python
my_matrix = found_com_relation(lp_coms.communities)
```

```python

import random

pos=nx.circular_layout(new_G)
plt.figure(num=None, figsize=(20, 20), dpi=500)
plt.axis('off')

nx.draw(new_G, pos)
edge_labels=dict([((u,v,),d['weight'])
             for u,v,d in new_G.edges(data=True)])

nx.draw_networkx_labels(new_G, pos, font_size=10, font_family='sans-serif')
nx.draw_networkx_edges(new_G, pos=pos)
nx.draw_networkx_edge_labels(new_G, pos , edge_labels=edge_labels,
                             label_pos=0.2, font_size=6)
plt.savefig("sonuclar/newGraf.png",figsize=(20, 20),  dpi=500)
```
